This notebook processes data from :

ICLR 2018: https://www.kaggle.com/krishnakalyan3/iclr-2018

ICLR 2017: https://www.kaggle.com/ahmaurya/iclr2017reviews

NIPS: https://www.kaggle.com/benhamner/nips-papers#paper_authors.csv

Into the following "Schema":

ID to Paper
[PAPERID, Paper]
id_to_paper

Paper to Author
[PAPERID, Author]
id_to_author

Venue
[PAPERID, Venue]
id_to_venue

YEAR
[PAPERID, Year]
id_to_year

ABSTRACT
[PAPERID, Abstract]
id_to_abstract

CATEGORY
[PAPERID, Category]
id_to_category

Paper to URL
[PAPERID, URL]
id_to_url

```
import pandas as pd
import hashlib
```

```
iclr_2018_df = pd.read_csv("iclr_2018.csv")
```

```
iclr_2018_df.head()
```

ID to PAPER:

```
print("Number of titles: " + str(iclr_2018_df['title'].count()))
print("Number of unique titles: :" + str(iclr_2018_df['title'].nunique()))
```

```
iclr_2018_df[iclr_2018_df.duplicated(keep=False, subset=['title', 'authors'])].sort_values(by=['title'])
```

```
iclr_2018_df.drop_duplicates(subset = ["title", "authors"], keep='first', inplace=True)
print("Number of titles: " + str(iclr_2018_df['title'].count()))
print("Number of unique titles: :" + str(iclr_2018_df['title'].nunique()))
```

```
id_to_paper = pd.DataFrame(columns=['ID', 'Title'])
id_to_paper['ID'] = iclr_2018_df.apply(lambda x: hashlib.sha256((str(x['title'])).encode()).hexdigest(), axis = 1)
id_to_paper['Title'] = iclr_2018_df['title']
id_to_paper.head()
```

```
for i in range(5):
  print(hashlib.sha256((str(iclr_2018_df['title'][i])).encode()).hexdigest())
```

```
print("Number of IDs: " + str(id_to_paper['ID'].count()))
print("Number of unique IDs: :" + str(id_to_paper['ID'].nunique()))
```

```
id_to_paper_dict = id_to_paper.set_index('Title').T.to_dict(orient='list')
print(id_to_paper_dict.get(iclr_2018_df['title'][0])[0])
```

PaperID to Author

```
id_to_author = pd.DataFrame(columns=['ID', 'Author'])
id_to_author['Author'] = iclr_2018_df.apply(lambda x: str(x["authors"]).split(" · "), axis = 1)
id_to_author['ID'] = iclr_2018_df.apply(lambda x: id_to_paper_dict.get(str(x["title"]))[0], axis = 1)
id_to_author.head()
```

```
id_to_author = id_to_author.explode('Author')
id_to_author.head()
```

ID To Venue

```
id_to_venue = pd.DataFrame(columns=['ID', 'Venue'])
id_to_venue['ID'] = id_to_paper['ID']
id_to_venue['Venue'] = "ICLR_2018"
id_to_venue.head()
```

ID to Year

```
id_to_year = pd.DataFrame(columns=['ID', 'Year'])
id_to_year['ID'] = id_to_paper['ID']
id_to_year['Year'] = 2018
id_to_year.head()
```

ID to Abstract

```
id_to_abstract = pd.DataFrame(columns=['ID', 'Abstract'])
id_to_abstract['ID'] = id_to_paper['ID']
id_to_abstract['Abstract'] = iclr_2018_df['abstract']
id_to_abstract.head()
```

ID to Category

```
id_to_category = pd.DataFrame(columns=['ID', 'Category'])
id_to_category['Category'] = iclr_2018_df.apply(lambda x: str(x["keywords"]).split(", "), axis = 1)
id_to_category['ID'] = iclr_2018_df.apply(lambda x: id_to_paper_dict.get(str(x["title"]))[0], axis = 1)
id_to_category = id_to_category.explode('Category')
id_to_category.head()
```

ID to URL

```
id_to_URL = pd.DataFrame(columns=['ID', 'URL'])
id_to_URL['ID'] = id_to_paper['ID']
id_to_URL['URL'] = iclr_2018_df['url']
id_to_URL.head()
```

ICLR 2017:


```
iclr_2017_df = pd.read_csv("iclr2017_papers.csv")
iclr_2017_df.head()
```

```
iclr_2017_df = iclr_2017_df[iclr_2017_df.decision != "Reject"]
iclr_2017_df.head()
```

```
print("Number of titles: " + str(iclr_2017_df['title'].count()))
print("Number of unique titles: :" + str(iclr_2017_df['title'].nunique()))
```

```
id_to_paper_iclr_2017 = pd.DataFrame(columns=['ID', 'Title'])
id_to_paper_iclr_2017['ID'] = iclr_2017_df.apply(lambda x: hashlib.sha256((str(x['title'])).encode()).hexdigest(), axis = 1)
id_to_paper_iclr_2017['Title'] = iclr_2017_df['title']
id_to_paper_iclr_2017.head()
```

```
id_to_paper_dict_iclr_2017 = id_to_paper_iclr_2017.set_index('Title').T.to_dict(orient='list')
print(id_to_paper_dict_iclr_2017.get(iclr_2017_df['title'][0])[0])
```

```
id_to_author_iclr_2017 = pd.DataFrame(columns=['ID', 'Author'])
id_to_author_iclr_2017['Author'] = iclr_2017_df.apply(lambda x: str(x["authors"]).split(", "), axis = 1)
id_to_author_iclr_2017['ID'] = iclr_2017_df.apply(lambda x: id_to_paper_dict_iclr_2017.get(str(x["title"]))[0], axis = 1)
id_to_author_iclr_2017.head()
```

```
id_to_author_iclr_2017 = id_to_author_iclr_2017.explode('Author')
id_to_author_iclr_2017.head()
```

```
id_to_venue_iclr_2017 = pd.DataFrame(columns=['ID', 'Venue'])
id_to_venue_iclr_2017['ID'] = id_to_paper_iclr_2017['ID']
id_to_venue_iclr_2017['Venue'] = "ICLR_2017"
id_to_venue_iclr_2017.head()
```

```
id_to_year_iclr_2017 = pd.DataFrame(columns=['ID', 'Year'])
id_to_year_iclr_2017['ID'] = id_to_paper_iclr_2017['ID']
id_to_year_iclr_2017['Year'] = 2017
id_to_year_iclr_2017.head()
```

```
id_to_abstract_iclr_2017 = pd.DataFrame(columns=['ID', 'Abstract'])
id_to_abstract_iclr_2017['ID'] = id_to_paper_iclr_2017['ID']
id_to_abstract_iclr_2017['Abstract'] = iclr_2017_df['abstract']
id_to_abstract_iclr_2017.head()
```

```
id_to_category_iclr_2017 = pd.DataFrame(columns=['ID', 'Category'])
id_to_category_iclr_2017['Category'] = iclr_2017_df.apply(lambda x: str(x["keywords"]).split(", "), axis = 1)
id_to_category_iclr_2017['ID'] = iclr_2017_df.apply(lambda x: id_to_paper_dict_iclr_2017.get(str(x["title"]))[0], axis = 1)
id_to_category_iclr_2017 = id_to_category_iclr_2017.explode('Category')
id_to_category_iclr_2017.head()
```

```
id_to_URL_iclr_2017 = pd.DataFrame(columns=['ID', 'URL'])
id_to_URL_iclr_2017['ID'] = id_to_paper_iclr_2017['ID']
id_to_URL_iclr_2017['URL'] = iclr_2017_df['pdf_link']
id_to_URL_iclr_2017.head()
```
